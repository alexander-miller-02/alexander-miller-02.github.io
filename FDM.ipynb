```
# Finite Difference Method
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

n = 1000

t = np.linspace(0,8,n)
y = np.sin(t)


# Construct FDM matrix
fdm_matrix = np.zeros((n,n))
h = t[1] - t[0]

central_coeff_positions= [-1, 0, 1]
central_coeff_values = [-0.5, 0, 0.5]

for j in range(-central_coeff_positions[0],n-central_coeff_positions[-1]):
  for i in range(len(central_coeff_positions)):
    fdm_matrix[j][j+central_coeff_positions[i]] = central_coeff_values[i]/h

forward_coeff_positions= [0, 1, 2]
forward_coeff_values = [-3/2, 2, -1/2]

for j in range(0, -central_coeff_positions[0]):
  for i in range(len(forward_coeff_positions)):
    fdm_matrix[j][j+forward_coeff_positions[i]] = forward_coeff_values[i]/h

reverse_coeff_positions= [-2, -1, 0]
reverse_coeff_values = [1/2, -2, 3/2]

for j in range(n-central_coeff_positions[-1], n):
  for i in range(len(reverse_coeff_positions)):
    fdm_matrix[j][j+reverse_coeff_positions[i]] = reverse_coeff_values[i]/h



fdm_matrix = fdm_matrix
dydt = np.matmul(fdm_matrix, y)

fig, ax = plt.subplots(figsize=(4,4))
sns.lineplot(x=t, y=y)
sns.lineplot(x=t, y=dydt)
```
