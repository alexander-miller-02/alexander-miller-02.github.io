```
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

fig, ax = plt.subplots(figsize=(5, 5))
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_aspect('equal')
line, = ax.plot([], [], color='blue')
title = ax.set_title('')

theta = np.linspace(0, 2 * np.pi, 100)

def init():
    line.set_data([], [])
    return line,

def update(frame):
    eccentricity = frame / 100 * 0.9
    x = np.cos(theta)
    y = np.sin(theta) * (1 - eccentricity)
    line.set_data(x, y)
    title.set_text(f'Eccentricity: {eccentricity:.2f}')
    return line, title

ani = FuncAnimation(fig, update, frames=101, init_func=init, blit=True, interval=50)

plt.close(fig)  # ðŸ’¡ Hide the static plot

HTML(ani.to_jshtml())
```
